```python
col = "city"
if col not in user_city.columns:
    raise KeyError(f"Missing column: {col}. Available: {list(user_city.columns)}")

# 取不重複值（保留原出現順序），NaN 會印成 <NA>
u = pd.unique(user_city[col].astype("object").where(~user_city[col].isna(), other=pd.NA))

print("\n" + "=" * 80)
print(f"Column: {col} | dtype: {user_city[col].dtype} | n_unique: {len(u)}")
print("-" * 80)
for v in u:
    print("<NA>" if pd.isna(v) else repr(v))

print("\nDONE.")
```

```python
import pandas as pd

s = user_city["city"]

counts = s.value_counts(dropna=False)
shares = s.value_counts(dropna=False, normalize=True)

summary = pd.DataFrame({
    "count": counts,
    "share": shares
})

# 想要百分比欄位（可選）
summary["share_pct"] = (summary["share"] * 100).round(2)

print(summary)
```
