```python
import pandas as pd

# 轉成 datetime（無法解析的會變 NaT）
rt = pd.to_datetime(trip["request_time"], errors="coerce", utc=True)

earliest = rt.min()
latest   = rt.max()

print("earliest request_time:", earliest)
print("latest   request_time:", latest)
```
