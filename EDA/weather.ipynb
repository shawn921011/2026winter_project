```python
import pandas as pd

# 如果 weather 還沒轉 datetime（保險起見）
weather["forecastDate"] = pd.to_datetime(weather["forecastDate"])

mask = (
    (weather["cityName"] == "南投縣") &
    (weather["forecastDate"].isin(pd.to_datetime(["2025-07-28"])))
)

# 印出全部欄位（避免被截斷）
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 200)

print(weather.loc[mask].sort_values(["forecastDate", "dayOrNight", "daypartName"]).to_string(index=False))
```
